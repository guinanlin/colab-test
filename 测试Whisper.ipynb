```python
import whisper
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import Audio, display
import time

# 加载tiny模型（最快的版本）
print("正在加载Whisper tiny模型...")
model = whisper.load_model("tiny")
print("模型加载完成！")

# 显示模型信息
print(f"模型大小: tiny (39M参数)")
print(f"预期速度: ~10x实时速度")
```

```python
import os
def transcribe_uploaded_file(file_path="/content/test.m4a"):
    """识别上传的音频文件"""

    # 检查文件是否存在
    if not os.path.exists(file_path):
        print(f"错误: 文件 {file_path} 不存在")
        print("请确保已上传音频文件到 /content/test.m4a")
        return None

    print(f"开始处理文件: {file_path}")
    print(f"文件大小: {os.path.getsize(file_path) / 1024 / 1024:.2f} MB")

    # 开始计时
    start_time = time.time()

    try:
        # 执行识别
        result = model.transcribe(
            file_path,
            language="zh",      # 指定中文提高速度
            fp16=False,         # CPU模式使用fp32
            verbose=True,       # 显示进度
            word_timestamps=True # 获取词级时间戳
        )

        end_time = time.time()
        duration = end_time - start_time

        # 显示结果
        print(f"\n=== 识别完成 ===")
        print(f"处理耗时: {duration:.2f}秒")
        print(f"检测语言: {result['language']}")
        print(f"识别文本: {result['text']}")


        return result

    except Exception as e:
        print(f"识别出错: {e}")
        return None

# 执行识别
result = transcribe_uploaded_file()
```
